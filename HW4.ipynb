```python
import matplotlib.pyplot as plt
import pandas as pd

housing_data = '/content/Data sets/boston_housing_data.csv'
df = pd.read_csv(housing_data)

plt.figure(figsize=(12, 8))

plt.plot(df.index, df['ZN'], label='ZN', color='green', linestyle='-')
plt.plot(df.index, df['INDUS'], label='INDUS', color='blue', linestyle='--')

plt.title("ZN and INDUS in Housing Data")
plt.xlabel("Sample Data Points")
plt.ylabel("Values")

plt.legend()

plt.style.use('bmh')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

col1 = [10, 15, 20, 25, 30]
col2 = [5, 10, 15, 20, 25]
categories = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(8, 6))
plt.bar(categories, col1, label='Column 1', color='blue', alpha=0.7)
plt.bar(categories, col2, label='Column 2', color='green', alpha=0.7, bottom=col1)
plt.title("Bar Chart with Large Title")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.legend(loc='lower left')
plt.show()

plt.figure(figsize=(8, 6))
plt.barh(categories, col1, label='Column 1', color='blue', alpha=0.7)
plt.barh(categories, col2, label='Column 2', color='green', alpha=0.7, left=col1)
plt.title("Horizontal Bar Chart with Large Title")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.legend(loc='upper right')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Data sets/drinks.csv')

x_variable = 'wine_servings'
y_variable = 'beer_servings'

plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_variable, y=y_variable, data=data, color='blue')
plt.title(f"Scatter Plot of {x_variable} vs {y_variable}")
plt.xlabel(x_variable)
plt.ylabel(y_variable)

plt.show()
```
